```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import netCDF4 as nc
import sys
import glob

sys.path.insert(1, '/home/jrichter/python/mqt/py_files')
from pip_tools_original_JackCopy import open_dist_data
from Dm_helpers import *
```

```python
# open PIP data for a given snow event
# get the 15-minute rolling averages of PSDs for each snowfall event --> see N0_lambda method
# perform Dm calculation on each of these rolling averages
```

```python
Dm_array_AR = get_Dm('/home/jrichter/python/mqt/txt_files/MQT_ARs_Snow.txt')

Dm_array_Deep = get_Dm('/home/jrichter/python/mqt/txt_files/Deep_NoAR_list.txt')
```

```python
Dm_array_AR_i = []
Dm_array_Deep_i = []

for i in range(Dm_array_AR.size):
    if not np.isnan(Dm_array_AR[i]):
        Dm_array_AR_i.append(Dm_array_AR[i])
    
for i in range(Dm_array_Deep.size):
    if not np.isnan(Dm_array_Deep[i]):
        Dm_array_Deep_i.append(Dm_array_Deep[i])
```

```python
plt.figure(dpi=150)
bin_D = np.arange(0.4, 15.2, 0.2)
plt.hist(Dm_array_AR_i, bins=bin_D, range=(0.4, 15), alpha=0.6, color="red", label="AR Snowfall")
plt.hist(Dm_array_Deep_i, bins=bin_D, range=(0.4, 15), alpha=0.6, color="blue", label="NoAR Snowfall")
plt.xlim(0.4, 15)
plt.ylim(0, 1500)
plt.title("PIP $D_m$ Histograms", size=20)
plt.xlabel("Mass-Weighted Mean Diameter ($D_m$) [mm]", size=14)
plt.ylabel("Counts", size=14)
plt.grid()
plt.legend()
#plt.savefig("/home/jrichter/python/mqt/Dm_histograms.png", bbox_inches=None, pad_inches=0.1)
plt.show()
```

```python
plt.figure(dpi=150)
bin_D = np.arange(0.4, 15.2, 0.2)
plt.hist(Dm_array_AR_i, bins=bin_D, color="red", histtype='step', label="AR Snowfall", linewidth=2.0)
plt.hist(Dm_array_Deep_i, bins=bin_D, color="blue", histtype='step', label="NoAR Snowfall", linewidth=2.0)
plt.xlim(0.4, 10)
#plt.yscale("log")
plt.title("PIP $D_m$ Histograms", size=20)
plt.xlabel("Mass-Weighted Mean Diameter ($D_m$) [mm]", size=14)
plt.ylabel("Counts", size=14)
plt.grid()
plt.legend()
#plt.savefig("/home/jrichter/python/mqt/Dm_histograms_step.png", bbox_inches=None, pad_inches=0.1)
plt.show()
```

```python
plt.figure(dpi=150)
bin_D = np.arange(0.4, 15.2, 0.2)
plt.hist(Dm_array_AR_i, bins=bin_D, color="red", histtype='step', label="AR Snowfall", linewidth=2.0)
plt.hist(Dm_array_Deep_i, bins=bin_D, color="blue", histtype='step', label="NoAR Snowfall", linewidth=2.0)
plt.xlim(0.4, 10)
plt.yscale("log")
plt.title("PIP $D_m$ Histograms", size=20)
plt.xlabel("Mass-Weighted Mean Diameter ($D_m$) [mm]", size=14)
plt.ylabel("Counts", size=14)
plt.grid()
plt.legend()
#plt.savefig("/home/jrichter/python/mqt/Dm_histograms_step_log.png", bbox_inches=None, pad_inches=0.1)
plt.show()
```
